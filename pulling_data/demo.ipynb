# Pulling Data For November and Saving to CSVs

```python
current=os.getcwd()#.replace('/','\\')
ticks=['SPY','^XSP','TSLA','NVDA','SPY','AAPL']
newpath = fr'{current}/options_data' 
for tick in ticks:
    if not os.path.exists(f'{newpath}/{tick}'):
        os.makedirs(f'{newpath}/{tick}')
```

```python
#another way to impute the maturity
import calendar
c = calendar.Calendar(firstweekday=calendar.SUNDAY)
maturity_month,maturity_year=11,2024
month,year = maturity_month,maturity_year

monthcal = c.monthdatescalendar(year,month)
third_friday = [day for week in monthcal for day in week if \
                day.weekday() == calendar.FRIDAY and \
                day.month == month][2]
print(f'the third friday in {month,year} is {third_friday}')
#print(type(third_friday))
```

```python
todays_date=datetime.today().strftime('%Y-%m-%d')
maturity=str(third_friday)
maturity_month,maturity_year=maturity.split('-')[1],maturity.split('-')[0]
ticks=['SPY','^XSP','TSLA','NVDA','SPY','AAPL']
for tick in ticks:
    print(tick)
    latest_data=pull_latest_data(tick,maturity)
    print(latest_data.shape)
    display(latest_data.head(5))
    dest=fr'options_data/{tick}/{tick}_{maturity_month},{maturity_year}_asOf{todays_date}.csv'
    #latest_data.to_csv(fr'{tick}_{maturity_month},{maturity_year}_asOf{todays_date}.csv')
    latest_data.to_csv(dest)
```
